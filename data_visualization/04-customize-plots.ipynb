````python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../datasets/mpg.csv')
````

<div>
<h2 class="breadcrumb">Combine Matplotlib, Pandas, and Seaborn plots in the same figure</h2><p>
</div>


````python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(7, 4), constrained_layout=True)
#matplotlib plot
ax1.scatter(df['displ'], df['hwy']);  
ax1.set_xlabel('displ'); 
ax1.set_ylabel('hwy');
#pandas plot
df.plot('displ', 'hwy', kind='scatter', ax=ax2);
#sns plot
sns.scatterplot(data=df, x='displ', y="hwy", ax=ax3);
sns.scatterplot(data=df, x='displ', y="hwy", hue='class', ax=ax4);
#Add figure title
fig.suptitle('Relationship between displacement and hwy mpg', fontsize=16);
````

- Use `subplot_mosaic` to  achieve more complex layouts, with Axes objects spanning columns or rows.



````python
fig, axs = plt.subplot_mosaic("""
    AB
    CD
    """, layout='constrained')
    
axs['A'].set_title('A')
axs['B'].set_title('B')
axs['C'].set_title('C');
axs['D'].set_title('D');
````

<h4>Tips</h4>

- A better way is to write a for loop to add values programmatically:



````python
fig, ax = plt.subplots()
sns.barplot(data=df, x='class', y='hwy', ax=ax);
for p in ax.patches:
    ax.annotate(
        text=round(p.get_height(),2), 
        xy=(p.get_x()+0.3, p.get_height()-3),
        color='b',
        bbox=dict(boxstyle="round,pad=0.3", fc="w", ec="w", lw=2)
    );
    # wayyyyy better 
````
