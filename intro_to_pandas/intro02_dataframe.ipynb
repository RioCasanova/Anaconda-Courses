```python
# picking or selecting columns and data
genre_year_topmovie = ["Genre", "Year", "Top Movie"]
movie_genres[genre_year_topmovie].head()
```

```python
# this returns a copy of the information in the chosen columns
# which is just another way of doing what was done above
movie_genres[["Genre", "Year"]].head()
```

```python
# this is evaluating the entire column by default, you can change this by changing axis=1
movie_genres[["Tickets Sold", "Top Movie Gross (That Year)"]].mean(axis=0)
```

```python
# when we change the axis=1 we get a row calculation that
# gets the average of the two rows that are specified
# this isn't a great dataset for this but it is just important to know how it workd
movie_genres[["Tickets Sold", "Top Movie Gross (That Year)"]].mean(axis=1)
```

```python
# this is returning the value that is in the column along with the number of times that it occurs
# it is only returning values that are unique (having no duplicates)
movie_genres["Movies Released"].value_counts()
```

```python
# in this next example we can use Trues as '1' and False as '0'
# and use the .sum() method to answer a question

# just outputting for my next step
movie_genres["Genre"].value_counts()
```

```python
# as you can see we have the same number of Comedys shown above
(movie_genres["Genre"] == "Comedy").sum()
```

```python
# percentage of the movies that are comedies
(movie_genres["Genre"] == "Comedy").mean()
```

```python
# you can also put conditional operations within the brackets to get a desired result
# you can also do this with an '&' expression
movie_genres[(movie_genres["Genre"] == "Documentary") | (movie_genres["Genre"] == "Reality")]
```

```python
# DataFrames work the same way as lists do in Python3
ogdf = pd.DataFrame({"x": [1, 2, 3]})
ogdf
```

```python
newdf = ogdf
ogdf["y"] = [4, 5, 6]
ogdf
```

```python
# even though we did not assign a 'y' to this copy of the DataFrame
# it still got the attributes. Be aware of this functionality.
newdf
```

```python
# so if we want to create a copy of a DataFrame we want to
# use the .copy() method
newdf = ogdf.copy()
ogdf["z"] = [6, 7, 8]
ogdf
```

```python
# note there is no further mutation on the copy
newdf
```

```python
# sorting the DataFrame
# you can sort by multiple columns by using square braces
# the inplace argument permanently changes the DataFrame to be sorted in that way if it is set to true
movie_genres.sort_values(by="Tickets Sold", ascending=False, ignore_index=True, inplace=True)
```
